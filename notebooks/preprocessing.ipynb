```python
import os
import pandas as pd
os.getcwd()  
```

```python
general_raw=pd.read_csv("..\data\general_matched.csv",encoding='utf-8')
old_raw=pd.read_csv("../data/old_matched.csv",encoding='utf-8')
```

```python
general_raw.head()
```

```python
general=general_raw.drop(columns='연번')
```

```python
time_cols = [c for c in general.columns if '시' in c]
general_melt=pd.melt(general,
                     id_vars=['날짜','호선','역번호','역명','구분'],
                     value_vars=time_cols,
                     var_name='시간대')
general_melt.head()
```

```python
old=old_raw.drop(columns='연번')
```

```python
old.head()
```

```python
time_cols = [c for c in old.columns if '시' in c]
old_melt=pd.melt(old,
                     id_vars=['날짜','역번호','역명','구분'],
                     value_vars=time_cols,
                     var_name='시간대')
old_melt.head()
```

```python
general_melt.drop(columns='호선')
```

```python
import numpy as np

general_melt['시간대_num'] = (
     general_melt['시간대']
    .str.extract(r'(\d{2})')   
    .astype(float)
)

# '06시 이전'만 예외 처리
general_melt.loc[general_melt['시간대'] == '06시 이전', '시간대_num'] = 5
```

```python
general_melt.head()
```

```python
general_melt = general_melt.drop(columns=['시간대'], errors='ignore')
general_melt= general_melt.rename(columns={'시간대_num': '시간대'})

# 3) 컬럼 순서 재배치
cols = ['날짜', '호선', '역번호', '역명', '구분', '시간대', 'value']
general_melt = general_melt[cols]
```

```python
import numpy as np

old_melt['시간대_num'] = (
    old_melt['시간대']
    .str.extract(r'(\d{2})')  
    .astype(int)
)

# 예외 처리
old_melt.loc[old_melt['시간대'] == '06시간대이전', '시간대_num'] = 5
old_melt.loc[old_melt['시간대'] == '24시간대이후', '시간대_num'] = 24
```

```python
old_melt = old_melt.drop(columns=['시간대'], errors='ignore')
old_melt= old_melt.rename(columns={'시간대_num': '시간대'})

# 3) 컬럼 순서 재배치
cols = ['날짜', '역번호', '역명', '구분', '시간대', 'value']
old_melt = old_melt[cols]
```

```python
old_melt
```

```python
df_final = pd.merge(general_melt, old_melt,
                     on = ['날짜','역번호','역명','구분','시간대'],
                     how = 'inner')
```

```python
df_final=df_final.rename(columns={
    'value_x':'general',
    'value_y':'old'
             
})
df_final.head()
```

```python
df_final.to_csv(
    'df_final.csv',
    index=False,
    encoding='utf-8-sig'
)
```
